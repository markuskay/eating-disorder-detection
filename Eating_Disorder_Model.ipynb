```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestClassifier

# Load dataset
df = pd.read_csv(r"C:\Users\HP\Desktop\FuWukari\Eating Disorder\Dataset.csv")

# -------------------------------
# STEP 1: BASIC CLEANING
# -------------------------------

df = df.drop_duplicates()
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

ordinal_map = {"Never":0, "Seldom":1, "Sometimes":2, "Often":3}
days_map = {"No days":0, "1-12 days":1, "13-27 days":2, "Every day":3}
target_map = {
    "Yes, I believe I do":1,
    "No, I donвЂ™t believe I do":0,
    "No, I don't believe I do":0
}

# Map ordinal fields
df = df.replace(ordinal_map)
df = df.replace(days_map)
df["Perception_EatingDisorder"] = df["Perception_EatingDisorder"].replace(target_map)

# Convert remaining non-numeric fields
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = le.fit_transform(df[col])

df = df.fillna(0)

# -------------------------------
# STEP 2: FEATURE ENGINEERING
# -------------------------------

# Identify thematic groups based on keyword patterns

restrictive_cols = [
    col for col in df.columns 
    if any(keyword in col for keyword in [
        "EatLess", "RefuseFood", "Avoid", "LimitFood", "Fasted"
    ])
]

emotional_cols = [
    col for col in df.columns 
    if "Eat_When" in col
]

reward_cols = [
    col for col in df.columns 
    if any(keyword in col for keyword in [
        "FoodTasty", "FoodSmells", "DeliciousFood", "DesireToBuy"
    ])
]

habit_cols = [
    col for col in df.columns 
    if any(keyword in col for keyword in [
        "Habit", "SpecificFoods", "AutomaticEating", "OutOfHabit"
    ])
]

body_cols = [
    col for col in df.columns 
    if col.startswith("Days_")
]

demographic_cols = [
    "Age_Range",
    "Gender",
    "Division_ofOrigin",
    "Current_EducationLevel",
    "Employment_Status",
    "Marital_Status"
]

# Create composite psychological scores

df["Restrictive_Eating_Score"] = df[restrictive_cols].mean(axis=1)
df["Emotional_Eating_Score"] = df[emotional_cols].mean(axis=1)

# Reward sensitivity (craving for tasty/attractive food)
reward_related = [
    col for col in df.columns 
    if "Tasty" in col or "Delicious" in col or "Smells" in col or "Desire" in col
]
df["Reward_Sensitivity_Score"] = df[reward_related].mean(axis=1)

df["Habitual_Eating_Score"] = df[habit_cols].mean(axis=1)
df["Body_Image_Distress_Score"] = df[body_cols].mean(axis=1)

# Compress demographic features into one score
df["Demographic_Score"] = df[demographic_cols].mean(axis=1)

# Days disturbance (overall body-image disturbance)
df["Days_Disturbance_Score"] = df[body_cols].sum(axis=1)

# -------------------------------
# STEP 3: SELECT MOST IMPORTANT FEATURES
# -------------------------------

target = "Perception_EatingDisorder"
engineered_features = [
    "Restrictive_Eating_Score",
    "Emotional_Eating_Score",
    "Reward_Sensitivity_Score",
    "Habitual_Eating_Score",
    "Body_Image_Distress_Score",
    "Demographic_Score",
    "Days_Disturbance_Score"
]

X = df[engineered_features]
y = df[target]

# Feature Importance
rf = RandomForestClassifier()
rf.fit(X, y)

importance_df = pd.DataFrame({
    "Feature": engineered_features,
    "Importance": rf.feature_importances_
}).sort_values(by="Importance", ascending=False)

print("\n=== FEATURE IMPORTANCE ===")
print(importance_df)

# Select top N features
top_features = importance_df["Feature"].head(7).tolist()

print("\n=== FINAL SELECTED FEATURES FOR MODEL ===")
print(top_features)

# -------------------------------
# STEP 4: BUILD FINAL REDUCED DATASET
# -------------------------------

final_df = df[top_features + [target]]

print("\n=== FINAL DATASET PREVIEW ===")
display(final_df.head())

print("\nOriginal Columns:", len(df.columns))
print("Reduced to:", len(final_df.columns)-1, "input features")
```

```python
# ============================================================
# 1. IMPORTS
# ============================================================
import pandas as pd
import numpy as np
import os
import joblib

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model

# ============================================================
# 2. LOAD DATASET
# ============================================================
df = pd.read_csv(r"C:\Users\HP\Desktop\FuWukari\Eating Disorder\Dataset.csv")
print("Raw Loaded Dataset:", df.shape)

df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# ============================================================
# 3. MAP RESPONSES в†’ NUMBERS
# ============================================================
ordinal_map = {"Never": 0, "Seldom": 1, "Sometimes": 2, "Often": 3}
days_map = {"No days": 0, "1-12 days": 1, "13-27 days": 2, "Every day": 3}
binary_map = {
    "Yes, I believe I do": 1,
    "No, I donвЂ™t believe I do": 0,
    "No, I don't believe I do": 0
}

df = df.replace(ordinal_map)
df = df.replace(days_map)
df["Perception_EatingDisorder"] = df["Perception_EatingDisorder"].replace(binary_map)

# ============================================================
# 4. FEATURE ENGINEERING (7 FEATURES)
# ============================================================
restrict_cols = [c for c in df.columns if "EatLess" in c or "Avoid" in c]
emotional_cols = [c for c in df.columns if "Eat_When" in c]
reward_cols = [c for c in df.columns if "Food" in c and "Eat_When" not in c]
habit_cols = [c for c in df.columns if "Habit" in c or "Automatic" in c]
body_cols = [c for c in df.columns if "Days_" in c]

df["Restrictive_Eating_Score"] = df[restrict_cols].mean(axis=1)
df["Emotional_Eating_Score"] = df[emotional_cols].mean(axis=1)
df["Reward_Sensitivity_Score"] = df[reward_cols].mean(axis=1)
df["Habitual_Eating_Score"] = df[habit_cols].mean(axis=1)
df["Body_Image_Distress_Score"] = df[body_cols].mean(axis=1)

df["ED_Risk_Score"] = df[
    [
        "Restrictive_Eating_Score",
        "Emotional_Eating_Score",
        "Reward_Sensitivity_Score",
        "Habitual_Eating_Score",
        "Body_Image_Distress_Score"
    ]
].mean(axis=1)

# Demographics
demo_cols = [
    "Age_Range", "Gender", "Division_ofOrigin",
    "Current_EducationLevel", "Employment_Status", "Marital_Status"
]

print("\nDemographic Columns Found:", demo_cols)

for col in demo_cols:
    df[col] = LabelEncoder().fit_transform(df[col])

df["Demographic_Score"] = df[demo_cols].mean(axis=1)

# ============================================================
# 5. CLINICAL TARGET LABEL
# ============================================================
df["ED_Risk_Label"] = (df["ED_Risk_Score"] >= 2.0).astype(int)
print("\nED Risk Label Counts:\n", df["ED_Risk_Label"].value_counts())

# ============================================================
# 6. FINAL FEATURE SET
# ============================================================
final_df = df[
    [
        "Restrictive_Eating_Score",
        "Emotional_Eating_Score",
        "Reward_Sensitivity_Score",
        "Habitual_Eating_Score",
        "Body_Image_Distress_Score",
        "ED_Risk_Score",
        "Demographic_Score",
        "ED_Risk_Label"
    ]
]

print("\nFinal Engineered Dataset:", final_df.shape)

# ============================================================
# 7. SCALING + SMOTE
# ============================================================
X = final_df.drop("ED_Risk_Label", axis=1)
y = final_df["ED_Risk_Label"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_scaled, y)

print("\nAfter SMOTE:", X_res.shape)

X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)

# ============================================================
# 8. MODEL TRAINING
# ============================================================
inputs = Input(shape=(7,))
x = Dense(32, activation="relu")(inputs)
x = Dense(16, activation="relu")(x)
x = Dropout(0.2)(x)
outputs = Dense(1, activation="sigmoid")(x)

model = Model(inputs, outputs)
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=40,
    batch_size=32,
    verbose=1
)

# ============================================================
# 9. EVALUATION
# ============================================================
y_pred = (model.predict(X_test) > 0.5).astype(int)

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, model.predict(X_test))
cm = confusion_matrix(y_test, y_pred)

print("\nACCURACY:", acc)
print("PRECISION:", prec)
print("RECALL:", rec)
print("F1 SCORE:", f1)
print("AUC:", auc)
print("\nCONFUSION MATRIX:\n", cm)

# ============================================================
# 10. SAVE MODEL + SCALER AUTOMATICALLY
# ============================================================
save_path = r"C:\Users\HP\Desktop\FuWukari\Eating Disorder"
os.makedirs(save_path, exist_ok=True)

model.save(os.path.join(save_path, "ed_risk_model.h5"))
joblib.dump(scaler, os.path.join(save_path, "ed_scaler.pkl"))

print("\nвњ” Model & Scaler Saved Successfully!")
print("Saved to:", save_path)
```
